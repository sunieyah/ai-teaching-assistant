## Problem Statement 1: Employee Performance Bonus Eligibility

```python
# Employee performance data
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Identify the highest performance score
highest_score = max(employees.values())

# Step 2: Find all employees with the highest score (handling ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Format and display the result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

## Problem Statement 2: Search Query Keyword Analysis

```python
# Input search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove common punctuation
import string
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split sentence into words
words = query.split()

# Step 4: Count frequency of each word
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Step 5: Filter words searched more than once
repeated_keywords = {word: count for word, count in word_count.items() if count > 1}

# Step 6: Display result
print(repeated_keywords)
```

## Problem Statement 3: Sensor Data Validation

```python
# Sensor readings data
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Initialize list to store valid readings
valid_readings = []

# Step 2: Loop through list with index
for hour, reading in enumerate(sensor_readings):
    
    # Step 3: Check if reading is even (valid)
    if reading % 2 == 0:
        valid_readings.append((hour, reading))

# Step 4: Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem Statement 4: Email Domain Usage Analysis

```python
# Email data
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Extract domains
domains = [email.split("@")[1] for email in emails]

# Step 2: Count frequency of each domain
domain_count = {}

for domain in domains:
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 3: Calculate percentage usage
total_emails = len(emails)

print("Email Domain Usage Analysis:")
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")


print("Extracted Domains:", domains)
print("Domain Count Dictionary:", domain_count)

```

## Problem Statement 5:Sales Spike Detection



```python
# Sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 2: Define spike threshold (30% above average)
threshold = average_sales * 1.3

# Step 3: Detect spike days
for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")


# Note:
# Based on strict 30% calculation, only Day 6 exceeds the threshold.
# The expected output includes Day 4 (2200), which is approximately
# 29.4% above average. The current solution follows the exact 30% rule.
```

## Problem Statement 6: Duplicate User ID Detection

```python
# User ID data
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count frequency of each user ID
user_count = {}

for user in user_ids:
    if user in user_count:
        user_count[user] += 1
    else:
        user_count[user] = 1

# Step 2: Identify duplicates (count > 1)
print("Duplicate User ID Detection:")

for user, count in user_count.items():
    if count > 1:
        print(f"{user} : {count} times")
```
